## **Ejercicio 1** (40%)

Construya un DataFrame a partir del siguiente diccionario

`exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'], 'score': [12.5, 9, 16.5, 10, 9, 20, 14.5, 12, 8, 19], 'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1], 'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}`

con las siguientes etiquetas para los índices

`labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']`

1. Seleccione solo las columnas *name* y *score*.
  
2. Filtre las filas donde el número de intentos (attemps) sea mayor que 2.

3. Filtre las filas donde el *score* se encuentre entre 15 y 20 (incluidos)

4. Cambie el *score* en la fila 'd' a 11.5

5. Agregue una nueva fila con etiqueta 'k' al dataframe con un valor que usted desee para cada columna

6. Borre la fila 'k' y entregue nuevamente el DataFrame original.

7. Ordene el DataFrame por nombre en orden ascendente.

8. Reemplace los valores de la columna *qualify* que contiene los valores *yes* y *no* por los valores booleanos True y False respectivamente.

```python
#Importando las librerias
import numpy as np
import pandas as pd
```

```python
#Se crea el DataFrame con los datos para el desarrollo de la prueba

#Se crea la serie para el indice
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

#Se crea el DataFrame
df = pd.DataFrame ({'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
                    'score': [12.5, 9, 16.5, 10, 9, 20, 14.5, 12, 8, 19],
                    'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
                    'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']},
                    index = labels)#En este punto se asigna el indice

#Se muestra el DataFrame
df
```

```python
#Ejercicio 1, Punto 1
#Se seleccionan solo las columnas name y score con iloc

df.iloc[ : , 0 : 2 ]
```

```python
#Ejercicio 1, Punto 2
#Se filtran las filas donde el número de intentos (attemps) sea mayor que 2

df.loc[df['attempts'] > 2]
```

```python
#Ejercicio 1, Punto 3
#Se filtran las filas donde el score se encuentre entre 15 y 20 (incluidos)

df.loc[(df['score'] >= 15) & (df['score'] <= 20)]
```

```python
#Ejercicio 1, Punto 4
#Se cambie el score en la fila 'd' a 11.5

df.loc['d' , 'score'] = 11.5
df
```

```python
#Ejercicio 1, Punto 5
#Se agrega una nueva fila con etiqueta 'k' al dataframe con valores deseados para cada columna

df = df.append(pd.Series({'name' : 'Valentina',
                          'score' : 10,
                          'attempts' : 1,
                          'qualify' : 'yes'},
                        name = 'k'))

#Se muestra el DataFrame
df
```

```python
#Ejercicio 1, Punto 6
#Se borra la fila 'k' y se entrega nuevamente el DataFrame original

df = df.drop('k')

#Se muestra el DataFrame original
df
```

```python
#Ejercicio 1, Punto 6
#Se ordena el DataFrame por nombre en orden ascendente por defecto

df = df.sort_values('name')

#Se muestra el DataFrame ordenado
df
```

```python
#Ejercicio 1, Punto 7
#Se reemplazan los valores de la columna qualify que contiene los valores yes y no por los valores booleanos True y False respectivamente

df['qualify'] = df['qualify'].map({'yes' : True, 'no' : False})

#Se muestra el DataFrame actualizado
df
```

## **Ejercicio 2** (30%)

Importe el dataset `athletes` del repositorio del curso y realice las siguientes tareas:

1. Importarlo como nombre athletes
2. Filtre el dataframe tal que solo se quede con los atletas que hayan ganado al menos una medalla.
3. Calcule la masa muscular por fila con una función como

$$I_{m.m.} = \frac{Weight}{Height^2}$$

4. Agrupe los datos por calculando la media por deporte para cada una del resto de características. ¿Cuál es el deporte en el que los jugadores tienen mayor peso promedio?

5. ¿Cuál es el pais con más participación femenina?

```python
#Ejercicio 2, Punto 1
#Se importan los datos con nombre athletes desde el repositorio del curso
athletes = pd.read_csv("https://raw.githubusercontent.com/tomasate/Datos_Clases/main/Datos_1/athletes.csv")

#Se muestran los datos
athletes
```

```python
#Ejercicio 2, Punto 2
#Se filtra el dataframe tal que solo queda con los atletas que hayan ganado al menos una medalla

athletes_ganadores = athletes.loc[(athletes['gold'] > 0) | (athletes['silver'] > 0) | (athletes['bronze'] > 0)]

#Se muestra el dataframe filtrado con los atletas que ganaron al menos una medalla
athletes_ganadores
```

```python
#Ejercicio 2, Punto 3
#Se calcula la masa muscular por fila con una función

#Se crea la nueva variable para la masa muscular mm
def calcular_Imm(athletes):
  """
  Esta función recibe como parámetro el dataframe para la lectura de los datos,
  retorna el índice de masa corporal a partir de su cálculo con la altura y el peso de cada atleta
  """
  #Aplicación de la fórmula para el cálculo del índice de masa corporal
  return (athletes['weight'] / (athletes['height']**2))

#Se llama la función y se aplica a nuestro DataFrame
athletes['Imm'] = athletes.apply(calcular_Imm, axis=1)

#Se muestra el dataframe con la nueva columna Imm
athletes
```

```python
#Ejercicio 2, Punto 4
#Se agrupan los datos por calculando la media por deporte para cada una del resto de características

#Se calcula la media
athletes_media = athletes.groupby('sport').mean()

#Se muestra el DataFrame con la media del resto de las caracteristicas
athletes_media

#¿Cuál es el deporte en el que los jugadores tienen mayor peso promedio?
#En el siguiente cuadro de código
```

```python
#¿Cuál es el deporte en el que los jugadores tienen mayor peso promedio?
#Se identifica el deporte en el que los jugadores tienen mayor peso promedio
deporte_mayor_peso_promedio = athletes_media['weight'].idxmax()

#Se muestra el deporte en el que los jugadores tienen mayor peso promedio
print(f"El deporte con mayor peso promedio es: {deporte_mayor_peso_promedio}, con un promedio de: {athletes_media['weight'].max()}")
```

```python
#Ejercicio 2, Punto 5
#¿Cuál es el pais con más participación femenina?

#Se hace el conteo para el sexo femenino por pais
conteo_paises_mujeres = athletes.loc[(athletes['sex'] == 'female')].groupby('nationality').sex.count()

#se identifica el pais con mayor participación femenina
pais_con_mas_mujeres = conteo_paises_mujeres.idxmax()

#Se muestra el pais en el que hay mayor participación fememina
print(f"El país con mayor participación femenina es: {pais_con_mas_mujeres}, con un total de {conteo_paises_mujeres.max()} mujeres")
```

## **Ejercicio 3 (30%)**

Importe el dataset de nba del repositorio de clase y responda (hint: Elimine las instancias con registros faltantes):

1. Cuáles son los equipos que pagan los salarios más altos (presente el resultado como un dataFrame con el nombre del equipo y la cantidad de dinero de la nómina)
2. Separe la variable *Age* en tres grupos y cree una nueva columna llamada *Age-Group* con esta información. Muestre un gráfico de barras con el número de jugadores por grupo de edad diferenciando en esta (mediante un código de color) la posición.
3. Construya una tabla dinámica que muestre el salario promedio de los jugadores por equipo y posición para para los cinco equipos que pagan mejores salarios.

```python
#Ejercicio 3
#Se importan los datos nba desde el repositorio del curso
nba = pd.read_csv("https://raw.githubusercontent.com/tomasate/Datos_Clases/main/Datos_1/nba.csv")

#Se eliminan las instancias con registros faltantes
nba = nba.dropna()

#Se muestran los datos
nba
```

```python
#Ejercicio 3, Punto 1
#Cuáles son los equipos que pagan los salarios más altos (presente el resultado como un dataFrame con el nombre del equipo y la cantidad de dinero de la nómina)

#Se agrupan y se suman los datos de salarios por equipos, y se ordenan de mayor a menor
equipos_salarios_altos = nba.groupby('Team').Salary.sum().sort_values(ascending = False)

#Se asignan los datos al DatFrame
equipos_nomina = pd.DataFrame({'Equipo' : equipos_salarios_altos.index,
                               'Nomina' : equipos_salarios_altos.values})

#Se muestra el nuevo DataFrame con los nombres de los equipos y la cantidad de dinero de la nómina
equipos_nomina
```

```python
#Ejercicio 3, Punto 2
#Separe la variable Age en tres grupos y cree una nueva columna llamada Age-Group con esta información
#Muestre un gráfico de barras con el número de jugadores por grupo de edad diferenciando en esta (mediante un código de color) la posición

#Identificamos la edad mínima y máxima
nba.Age.min()
nba.Age.max()
#La edad mínima es 19 y la edad máxima es 40

#Se definen los grupos de edad de: (19 - 25), (26 - 32), (33 - 40) por medio de la función
def categoria(nba):
  """
  Esta función recibe como parametro el DataFrame para la lectura de los datos,
  según la edad, identifica y clasifica el grupo de edad de los deportistas.
  """
  if nba['Age'] <= 25.0:
    return '(19 - 25)'
  if nba['Age'] >= 26.0 and nba['Age'] <= 32.0:
    return '(26 - 32)'
  if nba['Age'] >= 33.0 and nba['Age'] <= 40.0:
    return '(33 - 40)'

#Se crea la nueva variable Age-Group, se usa la función para la clasificación de los grupos de edad de cada deportista
nba['Age-Group'] = nba.apply(categoria, axis = 1)

#Se muestra el DataFrame con la clasificación de los grupos de edad
nba
```

```python
#Ejercicio 3, Punto 3
#Construya una tabla dinámica que muestre el salario promedio de los jugadores por equipo y posición para los cinco equipos que pagan mejores salarios

#Se identifican los índices de los primero 5 equipos que pagan mejores salarios
equipos_salarios_altos = nba.groupby('Team').Salary.sum().sort_values(ascending = False).head(5).index

#Se construye la tabla dinámica para los cinco equipos que pagan los mejores salarios,
#por equipo y posición con el salario promedio que es la función de agregación por defecto de la tabla dinámica
tabla_dinamica = pd.pivot_table(nba[nba['Team'].isin(equipos_salarios_altos)],
                                 values = 'Salary',
                                 index = 'Team',
                                 columns = 'Position')

#Se muestra la tabla dinámica
tabla_dinamica
```
